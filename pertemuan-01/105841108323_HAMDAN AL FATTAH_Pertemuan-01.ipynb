Tugas 1

```python
import numpy as np

# Mengatur seed agar hasil angka acak tetap konsisten
np.random.seed(42)

# Menghasilkan 20 angka acak
data_acak = np.random.rand(20) * 100

print("Daftar Angka Acak:")
print(data_acak.round(2))

print(f"\nRata-rata       : {np.mean(data_acak):.2f}")
print(f"Nilai Tengah    : {np.median(data_acak):.2f}")
print(f"Standar Deviasi : {np.std(data_acak):.2f}")
```

Tampil deretan 20 angka acak dengan format desimal yang rapi. Tepat di bawah deretan angka tersebut, ada rincian hasil perhitungan berupa Rata-rata, Nilai Tengah, dan Standar Deviasi.

Tugas 2


```python
import pandas as pd

# Penyusunan data nama mahasiswa dan skor ujian
np.random.seed(0)
data_akademik = {
    'Nama_Mahasiswa': [f"Mahasiswa_{i+1}" for i in range(10)],
    'Skor_Ujian': np.random.randint(50, 100, 10)
}

df_akademik = pd.DataFrame(data_akademik)

# Mengambil tiga baris dengan skor tertinggi
peringkat_atas = df_akademik.nlargest(3, 'Skor_Ujian')

print("Tabel Data Nilai Lengkap:")
print(df_akademik)

print("\nTiga Mahasiswa dengan Skor Tertinggi:")
print(peringkat_atas)
```

Muncul satu tabel penuh yang berisi nomor urut di sebelah kiri, diikuti oleh sepuluh nama mahasiswa, dan skor ujian mereka masing-masing. Di bagian bawahnya, tampil tabel kedua yang berukuran lebih ringkas, di mana isinya hanya memuat tiga baris nama mahasiswa dengan skor tertinggi.

Tugas 3

```python
import matplotlib.pyplot as plt

# Membuat pengaturan tata letak grafik berdampingan
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

# Grafik 1: Bar Chart
ax1.bar(df_akademik['Nama_Mahasiswa'], df_akademik['Skor_Ujian'], color='steelblue')
ax1.set_title('Perbandingan Nilai Mahasiswa')
ax1.set_ylabel('Skor')
ax1.tick_params(axis='x', rotation=45)

# Grafik 2: Histogram
ax2.hist(df_akademik['Skor_Ujian'], bins=5, color='coral', edgecolor='black')
ax2.set_title('Distribusi Frekuensi Nilai')
ax2.set_xlabel('Rentang Skor')
ax2.set_ylabel('Jumlah Mahasiswa')

plt.tight_layout()
plt.show()
```

Tampil dua area grafik berdampingan yang menyajikan representasi visual dari data nilai. Grafik batang menampilkan nilai spesifik tiap mahasiswa, sementara histogram menunjukkan seberapa banyak mahasiswa yang mendapatkan nilai dalam rentang tertentu. Pada bagian sumbu horizontal grafik kiri, label nama mahasiswa sengaja ditampilkan miring 45 derajat agar teks tidak saling tumpang tindih dan tetap mudah dibaca.

```python
from sklearn.datasets import load_wine

# Memuat dataset wine sebagai DataFrame
dataset_wine = load_wine(as_frame=True)
df_wine = dataset_wine.frame

print("Ukuran data:", df_wine.shape)
print("Kolom:", list(df_wine.columns))

print("\nInsight 1: Dataset Wine memiliki", df_wine.shape[0], "sampel dan", df_wine.shape[1] - 1, "fitur.")
print("Insight 2: Target memiliki", df_wine["target"].nunique(), "kelas unik.")
print("Insight 3: Rata-rata kandungan alcohol =", round(df_wine["alcohol"].mean(), 2))
```

Dalam output terdapat teks representasi jumlah baris dan kolom. Tepat di bawahnya, muncul rincian daftar nama kolom di dalam dataset. Setelah itu, baris teks Insight muncul menjelaskan jumlah sampel dan fitur yang dimiliki dataset. Teks Insight berikutnya menampilkan bahwa kolom Target memiliki 3 kelas berbeda. Pada baris terakhir, teks Insight menampilkan hasil perhitungan rata-rata kandungan alkohol di dalam dataset.
